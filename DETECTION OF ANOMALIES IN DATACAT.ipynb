# Mise en œuvre: mise à l'échelle des fonctionnalités

```python
#  Mettre à l'échelle les données en utilisant le logarithme naturel
log_data = np.log(data)

# Mettre à l'echelle les données de l'échantillon en utilisant le logarithme naturel
log_samples = np.log(samples)

```

# Mise en oeuvre: détection des valeurs aberrantes

```python
current_outliers = set()
multiple_outliers = set()

for feature in log_data.keys():
    
    # Calculez Q1 (25ème percentile des données) pour la fonction donnée
    Q1 = np.percentile(log_data[feature],25)
    
    #Calculez Q3 (75ème percentile des données) pour la fonction donnée.
    Q3 = np.percentile(log_data[feature],75)
    
    # Utilisez l'intervalle interquartile pour calculer une étape aberrante (1,5 fois l'intervalle interquartile).
    step = 1.5 * (Q3 - Q1)
    
    # Display les valeures aberrantes
    print("Data points considered outliers for the feature '{}':".format(feature))
    outlier_data = log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    for idx in outlier_data.index:
        if idx in current_outliers:
            multiple_outliers.add(idx)
        else:
            current_outliers.add(idx)
    display(outlier_data)
    
#FACULTATIF: Sélectionnez les index pour les points de données que vous souhaitez supprimer.
outliers  = [idx for idx in multiple_outliers]
print("Outliers ", outliers)

 # Supprimer les valeurs aberrantes, le cas échéant
good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)
```
