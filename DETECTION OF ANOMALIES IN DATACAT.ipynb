# Mise en oeuvre: dÃ©tection des valeurs aberrantes

```python
current_outliers = set()
multiple_outliers = set()

for feature in log_data.keys():
    
    # Calculez Q1 (25Ã¨me percentile des donnÃ©es) pour la fonction donnÃ©e
    Q1 = np.percentile(log_data[feature],25)
    
    #Calculez Q3 (75Ã¨me percentile des donnÃ©es) pour la fonction donnÃ©e.
    Q3 = np.percentile(log_data[feature],75)
    
    # Utilisez l'intervalle interquartile pour calculer une Ã©tape aberrante (1,5 fois l'intervalle interquartile).
    step = 1.5 * (Q3 - Q1)
    
    # Display les valeures aberrantes
    print("Data points considered outliers for the feature '{}':".format(feature))
    outlier_data = log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    for idx in outlier_data.index:
        if idx in current_outliers:
            multiple_outliers.add(idx)
        else:
            current_outliers.add(idx)
    display(outlier_data)
    
#FACULTATIF: SÃ©lectionnez les index pour les points de donnÃ©es que vous souhaitez supprimer.
outliers  = [idx for idx in multiple_outliers]
print("Outliers ", outliers)

 # Supprimer les valeurs aberrantes, le cas Ã©chÃ©ant
good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)
```
